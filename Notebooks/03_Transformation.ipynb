```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/dataset_clean_step1.csv")
```

Importation de notre dataset nettoyé

```python
df["Transaction Date"] = pd.to_datetime(
    df["Transaction Date"],
    errors="coerce"
)

```

Transformation de la colonne de date

```python
df["Year"] = df["Transaction Date"].dt.year
df["Month"] = df["Transaction Date"].dt.month
df["Day"] = df["Transaction Date"].dt.day
```

 Création de variables temporelles (année, mois, jour) à partir de la date  de transaction afin de faciliter les analyses temporelles


```python
df["Calculated Total"] = df["Quantity"] * df["Price Per Unit"]
```

 Calcul du montant total théorique afin de vérifier la cohérence entre la quantité, le prix unitaire et le montant total enregistré.
